### 1 task

```python
def split_n(number):
    l = []
    while number != 0:
        number, b = divmod(number, 10)
        l.append(b)
    return l[::-1]
```

```python
split_n(12345)
```

## 2 task

```python
def counter(number):
    l = split_n(number)
    result = {
        'evens': 0,
        'odds': 0
    }
    for i in l:
        key = 'evens' if i % 2 else 'odds'
        result[key] += 1
    return result
```

```python
counter(12345)
```

## 3 task

```python
def reverse_list_1_method(l):
    return l[::-1]

def reverse_list_2_method(l):
    i = len(l) - 1
    while i >= 0:
        yield l[i]
        i -= 1
```

```python
reverse_list_1_method([1,2,3,4,5])
```

```python
reverse_list_2_method([1,2,3,4,5])
```

## 6 task

```python
def count_duplicate_elements(l):
    return len([k for k, v in dict(Counter(l)).items() if v > 1])
```

```python
count_duplicate_elements([1,2,3,1,1,1,2])
```

```python
def delete_elements(l, condition):
    return [i for i in l if i != condition]
```

```python
def split_text_1_method(text):
    d = {}
    for word in re.sub('[^0-9a-z ]', '', text.strip().lower()).split():
        d[word] = d.get(word, 0) + 1
    return d

def split_text_2_method(text):
    return Counter(re.sub('[^0-9a-z ]', '', text.strip().lower()).split())
```

## 10 task

```python
def find_lines_1_method(list_lines, pattern):
    for line in list_lines:
        if pattern in line:
            yield line
    
def find_lines_2_method(list_lines, pattern):
    return [line for line in list_lines if pattern in line]
```

```python
for i in find_lines_1_method([
    'In publishing and graphic design',
    'Lorem ipsum is a placeholder text commonly',
    'used to demonstrate the visual form of a document',
    'or a typeface without relying on meaningful content.', 
    'Lorem ipsum may be used before final copy is available,', 
    'but it may also be used to temporarily replace copy',
    'in a process called greeking, which allows designers',
    'to consider form without the meaning of the text influencing the design.'
], ' copy'):
    print(i)
```

## 11 task

```python
def check_coordinates(list_coordinates):
    return [(latitude, longitude) for latitude, longitude in list_coordinates
           if latitude >= -90.0 and latitude <= 90.0 
            and longitude >= -180.0 and longitude <= 180.0]
```

```python
check_coordinates([(123.756, 89.3792), (-80.0456, 111.7489)])
```

## 12 task

```python
def balanced_paranthesis(text):
    brackets_stack = []
    brackets_dict = {
        'opening': {
            '{': '}',
            '[': ']',
            '(': ')'
        },
        'closing': {
            '}': '{',
            ']': '[',
            ')': '('
        }
    }
    for i, value in enumerate(text):
        if value in brackets_dict['closing'].keys():
            if not brackets_stack:
                return i
            if brackets_stack.pop()[1] != brackets_dict['closing'][value]:
                return i
        elif value in brackets_dict['opening'].keys():
            brackets_stack.append((i, value))
    if brackets_stack:
        return brackets_stack[0][0]
    else:
        return 'Success'
```

```python
balanced_paranthesis('[gfgv[[fcgvbh]]')
```

```python
balanced_paranthesis('[[[]]}]]')
```

```python
balanced_paranthesis('{vg[tesf(dfb)sg]gs}')
```
